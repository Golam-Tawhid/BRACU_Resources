```python
#Dictionary Tuple
#1
d1={"a": 100, "b": 100, "c": 200, "d": 300}
d2={"a": 300, "b": 200, "d": 400, "e": 200}
d3={}
lst1=[]
for i,j in d1.items():
  if i not in d3:
    d3[i]=j
for m,n in d2.items():
  if m not in d3:
    d3[m]=n
  else:
    d3[m]+=n
for i in d3.values():
  if i not in lst1:
    lst1.append(i)
for i in range (len(lst1)-1):
  for j in range(len(lst1)-1-i):
    if lst1[j]>lst1[j+1]:
      temp=lst1[j]
      lst1[j]=lst1[j+1]
      lst1[j+1]=temp
print(d3)
print("Values:",tuple(lst1)) 
```

```python
#2
def num_adder(min,max,div):
  sum=0
  for i in range(min,max):
    if i%div==0:
      sum+=i
  return sum
mini=int(input("Enter a minimum value: "))
maxi=int(input("Enter a maximum value: "))
divi=int(input("Enter a divisor: "))
print(num_adder(mini,maxi,divi))
```
