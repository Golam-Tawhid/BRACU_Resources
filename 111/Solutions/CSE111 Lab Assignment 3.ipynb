```python
#task1
class Calculator:
  def __init__(self,f_value,op,s_value):
    self.first_value=f_value
    self.operator=op
    self.second_value=s_value
    print("Let’s Calculate!")
    print("Value 1:",self.first_value)
    print("Operator:",self.operator)
    print("Value 2:",self.second_value)
    if self.operator=="+":
      self.add()
    elif self.operator=="-":
      self.subtract()
    elif self.operator=="*":
      self.multiply()
    elif self.operator=="/":
      self.divide()

  def add(self):
    result=(self.first_value)+(self.second_value)
    print("Result:",result)
  def subtract(self):
    result=(self.first_value)-(self.second_value)
    print("Result:",result)
  def multiply(self):
    result=(self.first_value)*(self.second_value)
    print("Result:",result)
  def divide(self):
    result=(self.first_value)/(self.second_value)
    print("Result:",result)
  
cal1=Calculator(int(input()),input(),int(input()))
```

```python
#task6
# Write your code here
class Calculator:
  def __init__(self):
    self.f=0
    self.s=0
    self.op=""
    self.val=0
    print("Calculator is ready!")
  def calculate(self,f,s,op):
    self.f=f
    self.s=s
    self.op=op
    if op=="+":
      val=f+s
      self.val=val
      return val
    elif op=="-":
      val=f-s
      self.val=val
      return val
    elif op=="*":
      val=f*s
      self.val=val
      return val
    elif op=="/":
      val=f/s
      self.val=val
      return val
  def showCalculation(self):
    print(f"{self.f}{self.op}{self.s}={self.val}")

c1 = Calculator()
print("==================")
val = c1.calculate(10, 20, '+')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 10, '-')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 5, '*')
print("Returned value:", val)
c1.showCalculation()
print("==================")
val = c1.calculate(val, 16, '/')
print("Returned value:", val)
c1.showCalculation()



```

```python
#task7
# Write your code here
class Student:
  def __init__(self,nm,id,dept,cg):
    self.name=nm
    self.id=id
    self.dept=dept
    self.cgpa=cg
  def calculate_CGPA(self):
    sum=0
    self.final=0
    for i in self.cgpa:
      sum+=i*3
    self.final=sum/(len(self.cgpa)*3)
    return self.final
  def print_details(self):
    print("Name:",self.name,",ID:",self.id)
    print("Department:",self.dept)
    print("CGPA:",self.final)
    r=""
    if self.final>3.80:
      r="Highest Distinction"
    elif self.final>3.65:
      r="High Distinction"
    elif self.final>3.50:
      r="Distinction"
    elif self.final>2.00:
      r="Satisfactory"
    if r!="":
      print(f"Your academic standing is '{r}'")
    if self.final<2.00:
      print("Sorry You Can’t Graduate")
  
s1 = Student('Dora', '15995599','CSE', [4,3.7,3.7,4])
s1.calculate_CGPA()
print("==========================")
s1.print_details()
print("==========================")
s2 = Student('Pingu', '12312322', 'EEE', [1.7,1.3,1.3,1.3,1])
s2.calculate_CGPA()
print("==========================")
s2.print_details()
print("==========================") 
s3 = Student('Bob', '13311331', 'CSE', [2,3,3,3.7,2.7,2.7])
s3.calculate_CGPA()
print("==========================")
s3.print_details()
```

```python
#task11
# Write your code here
class Spotify:
  def __init__(self,lst):
    self.lst=lst
    print("Welcome to Spotify!")
  def playing_number(self,nm):
    if nm>len(self.lst):
      return(f"{nm} number song not found. Your playlist has {len(self.lst)} songs only.\n##########################")
    else:
      return(f"Playing {nm} number song for you \nSong name: {self.lst[nm-1]} \n##########################")
  def add_to_playlist(self,new):
    self.lst.append(new)

user1 = Spotify(["See You Again", "Uptown Funk", "Hello"])
print("=========================")
print(user1.playing_number(4))
user1.add_to_playlist("Dusk Till Dawn")
print(user1.playing_number(3))
print(user1.playing_number(4))
```
