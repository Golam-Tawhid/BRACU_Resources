```python
#TASK_4
sitting_list = [10,30,20,70,11,15,22,16,58,100,12,56,70,80]

even_index_list=[]
odd_index_list=[]
final_list=[]

for count in range(0,len(sitting_list)):
  if count%2==0:
    even_index_list.append(sitting_list[count])
  else:
    odd_index_list.append(sitting_list[count])

for i in range (len(even_index_list)-1):
  for j in range(len(even_index_list)-1-i):
    if even_index_list[j]>even_index_list[j+1]:
      temp=even_index_list[j]
      even_index_list[j]=even_index_list[j+1]
      even_index_list[j+1]=temp

for i in range (len(odd_index_list)-1):
  for j in range(len(odd_index_list)-1-i):
    if odd_index_list[j]<odd_index_list[j+1]:
      temp=odd_index_list[j]
      odd_index_list[j]=odd_index_list[j+1]
      odd_index_list[j+1]=temp

for count in range(0,len(even_index_list)):
  final_list.append(even_index_list[count])
  final_list.append(odd_index_list[count])
print(final_list)
```

```python
#TASK_5
def my_sorting(my_dic):
  list_1=[]
  list_2=[]
  for k,v in my_dic.items():
    list_1.append(k)
    list_2.append(v)
  for i in range(len(list_2)):
    max_index=i
    for j in range(i,len(list_2)):
      if list_2[j]>list_2[i]:
        max_index=j
    temp_2=list_2[max_index]
    list_2[max_index]=list_2[i]
    list_2[i]=temp_2
    temp_1=list_1[max_index]
    list_1[max_index]=list_1[i]
    list_1[i]=temp_1
  return list_1

lst = [ ["Alan", 95, 87, 91], ["Turing", 92, 90, 83], ["Elon", 87, 92, 80], ["Musk", 85, 94, 90] ]
new_dic={}
for elm in lst:
  new_dic[elm[0]]=elm[1:]
course=input("Enter the course name: ")
my_list=["CSE110","PHY111","MAT110"]
count=0

if course in my_list:
  if course=="CSE110":
    count=1
  elif course=="PHY111":
    count=2
  elif course=="MAT110":
    count=3
  for elm in lst:
    new_dic[elm[0]]=elm[count]
  final_list=my_sorting(new_dic)
  for i in final_list:
    print(i)
else:
  print("Invalid course")
```

```python
#TASK_9
import math
points = [(1,7), (4,5), (-1,7), (-2,0), (1,1), (5,-1)]
min_dist=math.inf
min_pnt=()
for i in points:
  dist=math.sqrt(i[0]**2 + i[1]**2)

  if dist<min_dist:
    min_dist=dist
    min_pnt=i
print("Minimum distance = ",min_dist)
print(f"Here the closest point is {min_pnt} which has a distance of {min_dist} from the origin.")
```
