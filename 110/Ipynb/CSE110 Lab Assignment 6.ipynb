```python
#11
def rem_duplicate(tup):
  my_list=[]
  for i in tup:
    if i not in my_list:
      my_list.append(i)
  my_tup=tuple(my_list)
  return my_tup
print(rem_duplicate(("Hi", 1, 2, 3, 3, "Hi",'a', 'a', [1,2])))
```

```python
#12
def function_name(my_list):
  final_list=[]
  count=0
  for i in my_list:
    if final_list.count(i)<2:
      final_list.append(i)
    else:
      count+=1
  print("Removed:",count)
  return final_list
print(function_name([1, 2, 3, 3, 3, 3, 4, 5, 8, 8]))
```

```python
#13
my_operator=input()
operand_1=float(input())
operand_2=float(input())
def function_name(sym,num_1,num_2):
  sum=0
  if sym=="+":
    sum=num_1+num_2
  elif sym=="-":
    sum=num_1-num_2
  elif sym=="*":
    sum==num_1*num_2
  elif sym=="/":
    sum=num_1/num_2
  else:
    return "Invalid Operator"
  return float(sum)
print(function_name(my_operator,operand_1,operand_2))
```

```python
#14
word=input("Enter word: ")
pos=int(input("Enter position: "))

def function_name(n_word,position):
  new_str=n_word[0]
  last=" "
  for i in range(1,len(n_word)):
    if i%position!=0:
      new_str=new_str+n_word[i]
    else:
      last=last+n_word[i]
  add=new_str+last
  return add
print(function_name(word,pos))
```
